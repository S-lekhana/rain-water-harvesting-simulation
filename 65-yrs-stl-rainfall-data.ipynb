# 65 Years of St. Louis Rainfall Data

## Dataset

The following data was requested from NOAA:
https://www.ncdc.noaa.gov/cdo-web/search

|Field|Value|
|-----|-----|
|**Start Date**|1948-08-01|
|**End Date**|2013-12-23|
|**Locations**|St. Louis, MO US (CITY:US290021)|
|**Custom Flag(s)**|Station Name, Geographic Location, Include Data|
|**Units**|Standard|
|**Data Types**|HPCP - Precipitation|

```python
import pandas as pd

df = pd.read_csv('./2261150.csv', parse_dates=[5])
df
```

```python
df.groupby('STATION_NAME').agg(['min', 'max']).sort_values(by=[('DATE', 'min'), ('DATE', 'max')], ascending=True)
```

```python
mm = df.groupby('STATION_NAME').agg(['min', 'max'])
three_sixty_five_days = pd.Timedelta(value=365, unit='days')
rng = (mm['DATE']['max'] - mm['DATE']['min']) / three_sixty_five_days
rng.sort_values(ascending=False)
```
